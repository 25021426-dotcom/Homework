Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input())

for i in range(1, h + 1):
    # in khoảng trắng bên trái
    print(" " * (h - i), end="")

    if i == 1:
        # đỉnh tam giác
        print("*")
    elif i == h:
        # đáy tam giác
        print("*" * (2 * h - 1))
    else:
        # 2 cạnh bên
        print("*" + " " * (2 * i - 3) + "*")
```

Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python

def tim_uoc(n):
    u=[]
    for i in range(1,n+1):
        if n%i==0:
            u.append(i)
    return u 


def uoc_chung(a,b):
    u1=tim_uoc(a)
    u2=tim_uoc(b)
    c=[]
    for i in u1:
        if i in u2:
            c.append(i)
    return c

a,b=map(int, input().split())
print(uoc_chung(a,b))
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
def tim_uoc(n):
    u=[]
    for i in range(1,n):
        if n%i==0:
            u.append(i)
    return u 
a=int(input())
c=tim_uoc(a)
s=0
for i in c:
    s+=i
if s==a:
    print(a,'là số hoàn hảo')
else:
    print(a,'không là số hoàn hảo')
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n

```python
n=int(input())


def tong(k):
    s=0
    for i in range(1,k+1):
        s+=i
    return s

k=0
while tong(k)<n:
    k=k+1
print(k-1)
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
while True:
    A=int(input())
    if A>0:
        break

def tong_phan_so(n):
    s=0
    for i in range(1,n+1):
        s+=1/i
    return(s)

k=0
while tong_phan_so(k)<A:
    k=k+1

print(k)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
x1=True
u=[]

while True:
    x = int(input())
    u.append(x)
    if x == -1:
        break
    
    if x1:
        max=x
        min=x
        x1=False
    else:
        if x>max:
            max=x
        elif x<min:
            min=x

print(u)
print(max,min)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không

```python
def tim_uoc(n):
    u=[]
    for i in range(2,n):
        if n%i==0:
            u.append(i)
    return u 
n=int(input())
u=[]
for i in tim_uoc(n):
    if i%3==0:
        u.append(i)
if u==tim_uoc(n):
    print(n,'có dạng 3^k')
else:
    print(n,'không có dạng 3^k')
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A



```python
A=int(input())
def fib(n):
    f=[0,1]
    for i in range(2,n+1):
        f.append(i)
        f[i]=f[i-1]+f[i-2]
    return f[n]

n=2
while fib(n)<A:
    n=n+1

print(fib(n))

print(n-1)
```
