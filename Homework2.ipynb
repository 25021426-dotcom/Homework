Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b, c=map(float, input().split())


p=a+b+c
k=p/2
s=(k*(k-a)*(k-b)*(k-c))**0.5


if a+b<c or a+c<b or b+c<a:
    print('khong la ba canh cua mot tam giac')

else:
    if a==b==c:
        print('la tam giac deu') 

    elif (a**2)+(b**2)-(c**2)<0.0001 or (c**2)+(b**2)-(a**2)<0.0001 or (a**2)+(c**2)-(b**2)<0.0001:
        if a==b or b==c or a==c:
            print('la tam giac vuong can')
        else:
            print('la tam giac vuong')

    elif a==b or b==c or a==c:
        print('la tam giac can')

    else:
        print('la tam giac thuong')

print(f'chu vi {p:.2f}')
print(f'dien tich {s:.2f}')
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c=map(float, input().split())
delta=(b**2)-4*a*c
if delta>=0:
    x1=(-b-(delta**0.5))/2*a
    x2=(-b+(delta**0.5))/2*a
    if x1==x2:
        print(f'phương trình có nghiệm duy nhất {x1:.2f}')
    else:
        print(f"phương trình có nghiệm là {x1:.2f} và {x2:.2f}")
else :
    print('phương trình vô nghiệm')
```

Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).

Bài 8: Giải hệ phương trình tuyến tính:
(ax + by = m)
(cx + dy = n)

```python
a,b,m=map(float, input().split())
c,d,n= map(float, input().split())
D=a*d-c*b
dx=m*d-n*b
dy=a*n-c*m
if D!=0:
    x=dx/D
    y=dy/D
else:
    if dx==0 and dy==0:
        print('Hệ vô số nghiệm')
    else:
        print('Hệ vô nghiệm')
    
print(f'x là {x:.2f}')
print(f'y là {y:.2f}')
```

ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao
