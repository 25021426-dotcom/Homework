Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def Amstrong(n):
    k=n*10
    u=0
    d=len(list(str(n)))
    while True:
        k=k//10
        t=k%10
        if t==0:
            break
        u+=(t**d)
    return u==n
n=int(input())
print(Amstrong(n))
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a,b,c):
    delta=(b**2)-4*a*c
    if delta<0:
        return 'Vô Nghiệm'
    elif delta==0:
        x=-b/2*a
        return f'{x:.2f}'
    else:
        x1=(-b+delta**0.5)/2*a
        x2=(-b-delta**0.5)/2*a
        return f'{x2:.2f},{x1:.2f}'
    
a,b,c=map(int, input().split())
print(quadratic_roots_count(a,b,c))
    
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_colinear(x1,y1,x2,y2,x3,y3):
    Ax=x1-x2
    Ay=y1-y2
    Bx=x2-x3
    By=y2-y3
    return Ax*By==Ay*Bx

x1,x2,x3,y1,y2,y3=map(int,input().split())
print(are_colinear(x1,y1,x2,y2,x3,y3))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3)

```python
def goc(x1,x2,x3,y1,y2,y3):
    import math
    a1=(x1-x2,y1-y2)
    a2=(x3-x2,y3-y2)
    ab=((a1[0]**2)+(a1[1]**2))**0.5
    bc=((a2[0]**2)+(a2[1]**2))**0.5
    cos=(a1[0]*a2[0]+a1[1]*a2[1])/(bc*ab)
    rad=math.acos(cos)
    do=math.degrees(rad)
    
    return do
x1=3
y1=0
x2=0
y2=0
x3=0
y3=7
print(goc(x1,x2,x3,y1,y2,y3))
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def to_nho(c):
    max=c[0]
    min=c[0]
    for i in c:
        if i>max:
            max=i
        elif i<min:
            min=i
    
    return min,max
c=list(map(int,input().split()))
print(to_nho(c))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
def giai_thua(n):
    t=1
    for i in range(1,n+1):
        t*=i
    return t

def he_so_C(n,k):
    he_so=(giai_thua(n))/((giai_thua(k))*(giai_thua(n-k)))
    return he_so

n=1
k=2
print(he_so_C(n,k))
```

W5A31 Viết hàm tính tổng bình phương các số từ 1 đến n.
