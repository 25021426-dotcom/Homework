W4A20 :Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n=int(input())
t=1
while True:
    t=2*t
    if t==n:
        print(n,"có dạng 2^k")
        break
    else:
        if t>n:
            print(n,'Không có dạng 2^k')
            break
        else:
            continue



    
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
def fib(n):
    f=[0,1]
    for i in range(2,n+1):
        t=f[i-1]+f[i-2]
        f.append(t)
    return t

print(fib(11))
    
```

27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
while True:
    s=str(input())

    if all(c.isalpha() or c == ' ' for c in s):
          break
    else:
        continue





t = 0
in_word = False

for c in s:
    if c != ' ':
        if not in_word:
            t += 1
            in_word = True
    else:
        in_word = False

print('số từ trong (', s, ') là', t)
```

Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
while True:
    s=str(input())

    if all(c.isalpha() or c == ' ' for c in s):
          break
    else:
        continue

u=''
s=s.strip()
for i in s:
    u+=i
    if i==' ':
         break
    
print(u)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
s=str(input())
t=0
c=0
for i in s:
    k=ord(i)
    if k in range(65,91):
        t+=1
    elif k in range(97,123):
        c+=1

print(s,t,c)
```
